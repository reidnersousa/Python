```python
import pandas as pd


numero_1 = [11,2,10,7,24]
numero_2 = [25,20,16,15,29]
numero_3 = [26,29,22,16,40]
numero_4 = [59,33,69,61,68]
numero_5 = [68,46,76,62,76] 
ganhador = ["sim","sim","sim","sim","sim"]
dados = {
 'numero_1':numero_1,
 'numero_2':numero_2,
 'numero_3':numero_3,
 'numero_4':numero_4,
 'numero_5':numero_5,
 'v':ganhador
  }
pessoas = pd.DataFrame(dados)
display(pessoas)
#pessoas.to_csv('pessoas.csv')

print(type(pessoas))
```

```python
from random import randrange, uniform


for i in range(10):
  nova_linha=[randrange(1,80),randrange(1,80),randrange(1,80),randrange(1,80),randrange(1,80),"nao"]
  pessoas=pessoas.append(pd.Series(nova_linha, index=pessoas.columns[:len(nova_linha)]), ignore_index=True)

```

```python
pessoas['v']=pessoas['v'].replace('sim',0)
pessoas['v']=pessoas['v'].replace('nao',1)

pessoas
```

```python
y = pessoas['v']
x = pessoas.drop('v',axis = 1)

```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino,  y_teste=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier(n_estimators=100)

modelo.fit(x_treino,y_treino)

resultado = modelo.score(x_teste,y_teste)
print("Acurácia",resultado)
```
