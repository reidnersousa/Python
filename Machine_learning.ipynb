```python
def escolha_Arquivo(escolha):
    
    switcher ={
        0:"https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"   
    }
    return switcher.get(escolha,"escolha invalida")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import seaborn as sns
from pprint import pprint
from pprint import pprint

url_path =escolha_Arquivo(0) ### escolhe qual datecenter vai puxa o grafo


df = pd.read_csv(url_path )

df.head()
```

```python
display(df)
```

convetendo as variaveis string num inteiro 

```python
df['species']=df['species'].replace('setosa',0)
df['species']=df['species'].replace('versicolor',1)
df['species']=df['species'].replace('virginica',2)

```

separando as variáveis entre preditores e variável alvo

```python
y = df['species']
x = df.drop('species',axis = 1)

print("tamanho do database",df.shape,"y tabela so com species",y.shape," tabela sem a coluna species",x.shape)

```

separandos os dados 

```python
from sklearn.model_selection import train_test_split



x_treino, x_teste, y_treino,  y_teste=train_test_split(x,y,test_size=0.30)
```

Treinando

```python
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier(n_estimators=100)

modelo.fit(x_treino,y_treino)

resultado = modelo.score(x_teste,y_teste)
print("Acurácia",resultado)
```
